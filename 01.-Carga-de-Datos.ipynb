## Carga de Datos

```python
import pandas as pd
import numpy as np
import re
import warnings

warnings.filterwarnings('ignore')
```

```python
df_train = pd.read_csv('../NR - Data/titanic-train.csv')

df_train.head()
```

```python
df_train.shape
```

```python
df_test = pd.read_csv('../NR - Data/titanic-test.csv')

df_test.head()
```

```python
df_test.shape
```

## Feature Engineering

```python
list_df = [df_train,df_test]
```

### Longitud del Nombre

```python
for df in list_df:
    df['Name_length'] = df['Name'].apply(len)

#df_train['Name_length'] = df_train['Name'].apply(len)
#df_test['Name_length'] = df_test['Name'].apply(len)
```

```python
df_train.head()
```

```python
df_test.head()
```

### ¿Tiene Cabina?

```python
df_train['Cabin'].unique()
```

```python
for df in list_df:
    df['Has_Cabin'] = df['Cabin'].apply(lambda x: 0 if type(x) == float else 1)
```

```python
df_train.head()
```

### Tamaño de la Familia

```python
for df in list_df:
    df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
df_train.head()
```

### ¿Está Solo? 

```python
for df in list_df:
    df['IsAlone'] = 0
    df.loc[df['FamilySize'] == 1,'IsAlone'] = 1
```

```python
df_train.head()
```

### Eliminar NaN's  de la Columna 'Embarked' 

```python
df_train['Embarked'].unique()
```

```python
for df in list_df:
    df['Embarked'] = df['Embarked'].fillna('S')
```

```python
df_train['Embarked'].isna().sum()
```

### ¿Qué tipo de billete tiene? 

```python
for df in list_df:
    df['Fare'] = df['Fare'].fillna(df_train['Fare'].median())
```

```python
df_train['CategoricalFare'] = pd.qcut(df_train['Fare'],4)
```

```python
df_train['CategoricalFare'].unique()
```

### Variable Categórica de Edad 

```python
for df in list_df:
    age_avg = df['Age'].mean()
    age_std = df['Age'].std()
    age_null_count = df['Age'].isnull().sum()
    age_null_random_list = np.random.randint(age_avg - age_std,age_avg + age_std,size = age_null_count)
    df['Age'][np.isnan(df['Age'])] = age_null_random_list
    df['Age'] = df['Age'].astype(int)
```

```python
df_train.head()
```

```python
df_train['CategoricalAge'] = pd.cut(df_train['Age'],5)
```

```python
df_train['CategoricalAge'].unique()
```

### Estatus Social 

```python
df_train['Name'].unique()
```

```python
def get_title(name):
    title_search = re.search(' ([A-z]+)\.',name)
    if title_search:
        return title_search.group(1)
    return ""
```

```python
for df in list_df:
    df['Title'] = df['Name'].apply(get_title)
```

```python
df_train['Title'].unique()
```

```python
for df in list_df:
    df['Title'] = df['Title'].replace(
        ['Lady','Countess','Capt','Col','Don','Dr','Major','Rev','Sir','Jonkheer','Dona'],'Rare')
    df['Title'] = df['Title'].replace('Mlle','Miss')
    df['Title'] = df['Title'].replace('Ms','Miss')
    df['Title'] = df['Title'].replace('Mme','Mrs')
```

```python
df_train['Title'].unique()
```

 ## Mapping 

### 'Sex' 

```python
for df in list_df:
    df['Sex'] = df['Sex'].map({'female':0,'male':1}).astype(int)
```

```python
df_train.head()
```

### 'Title' 

```python
for df in list_df:
    df['Title'] = df['Title'].map({'Mr':1,'Miss':2,'Mrs':3,'Master':4,'Rare':5}).astype(int)
    df['Title'] = df['Title'].fillna(0).astype(int)
```

```python
df_train.head()
```

### 'Embarked' 

```python
for df in list_df:
    df['Embarked'] = df['Embarked'].map({'S':0,'C':1,'Q':2}).astype(int)
```

```python
df_train.head()
```

### 'Fare'

```python
df_train['CategoricalFare'].unique()
```

```python
for df in list_df:
    df.loc[df['Fare'] <= 7.91,'Fare'] = 0
    df.loc[(df['Fare'] > 7.91) & (df['Fare'] <= 14.454),'Fare'] = 1
    df.loc[(df['Fare'] > 14.454) & (df['Fare'] <= 31),'Fare'] = 2
    df.loc[ df['Fare'] > 31, 'Fare'] = 3
    df['Fare'] = df['Fare'].astype(int)
```

```python
df_train.head()
```

### 'Age'

```python
df_train['CategoricalAge'].unique()
```

```python
for df in list_df:
    df.loc[df['Age'] <= 16,'Age']= 0
    df.loc[(df['Age'] > 16) & (df['Age'] <= 32),'Age'] = 1
    df.loc[(df['Age'] > 32) & (df['Age'] <= 48),'Age'] = 2
    df.loc[(df['Age'] > 48) & (df['Age'] <= 64),'Age'] = 3
    df.loc[df['Age'] > 64,'Age'] = 4    
```

```python
df_train.head()
```

## Feature Selection 

```python
drop_elements = ['PassengerId','Name','Ticket','Cabin','SibSp']

df_train = df_train.drop(drop_elements,axis = 1)
df_train = df_train.drop(['CategoricalAge','CategoricalFare'],axis = 1)
```

```python
df_train.head()
```

```python
ids = df_test['PassengerId']
```

```python
df_test  = df_test.drop(drop_elements,axis = 1)
```

```python
df_test.head()
```
